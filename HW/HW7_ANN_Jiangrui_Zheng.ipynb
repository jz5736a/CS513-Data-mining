```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
# import the dataset
# Making a list of missing value types
missing_values = ["?", ""]
df = pd.read_csv('wisc_bc_ContinuousVar.csv', na_values = missing_values) 
print(df.shape)
df.head(10)
```

```python
df.isnull().sum()
```

# MinMax

```python
X = df.drop(['id', 'diagnosis'], axis=1)
y= df['diagnosis']

scaled_features = MinMaxScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_features, y, test_size=0.3, random_state=1)    # Use 30% test 70% training data.
```

```python
print("X_train Shape : ", X_train.shape)
print("X_test Shape  : ", X_test.shape)
print("y_train Shape : ", y_train.shape)
print("y_test Shape  : ", y_test.shape)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, recall_score, precision_score

mlpc_model = MLPClassifier(hidden_layer_sizes=5,random_state=1) # five (5) nodes in the hidden layer
mlpc_model.fit(X_train, y_train)
y_pred = mlpc_model.predict(X_test)

print('Accuracy Score: {:.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
