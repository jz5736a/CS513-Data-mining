III.	Replacing the numerical missing values with the “mode” of the corresponding columns

```python
# Replacing the missing values with the “mode”(most frequent value) of the column
# We found that only MonthlyIncome column contains missing values
data['MonthlyIncome'] = data['MonthlyIncome'].fillna(data['MonthlyIncome'].mode()[0])
```
